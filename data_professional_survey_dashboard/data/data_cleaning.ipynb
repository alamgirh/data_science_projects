```python
df = df.drop(columns=['Email'])
df.head(4)
```

```python
cols = ['Q1 - Which Title Best Fits your Current Role?', 
        'Q4 - What Industry do you work in?', 
        'Q5 - Favorite Programming Language',
        'Q8 - If you were to look for a new job today, what would be the most important thing to you?',
       'Q11 - Which Country do you live in?']

for col in cols:
    for k, item in enumerate(df[col]):
        if 'Other' in item:
            df[col][k] = 'Others'

        
```

```python
df.head(5)
```

Convert Q3 - Current Yearly Salary (in USD) to average salary

```python
length = len(df['Q3 - Current Yearly Salary (in USD)'])
for i in range(0, length):
    salary = df['Q3 - Current Yearly Salary (in USD)'][i].split('-')
    #print(len(salary), i)
    for k in range(0,len(salary)):
        if salary[k] == '0':
            salary[k] = int(salary[k])
        else: 
            if salary[k][-1] == 'k':
                salary[k] = int(salary[k][:-1])
            else:
                salary[k] = int(salary[k][:-2])
    salary_ave = sum(salary)/len(salary)        
    df['Q3 - Current Yearly Salary (in USD)'][i] = salary_ave
```

```python
df.rename({'Q3 - Current Yearly Salary (in USD)': 'Q3 - Average Salary (in USD)'}, axis=1, inplace=True)
```
